```python
import pandas as pd
import unicodedata
```

```python
# Green areas (xlsx)
df_green = pd.read_excel('data/Share_of_green_areas_and_green_area_per_capita_in_cities_and_urban_areas_1990_-_2020.xlsx', sheet_name='Data')
# print(df_green.head(10))

df_green.to_csv("data/green_areas_data.csv", index=False)

print("Green areas columns:", df_green.columns)
print("Sample cities:", df_green['City Name'].unique()[:10])
print(len(df_green))
```

```python
# Pollution (csv)
df_pollution = pd.read_csv('data/global_air_pollution_data.csv')
print("Pollution columns:", df_pollution.columns)
print("Sample cities:", df_pollution['city_name'].unique()[:10])
print(len(df_pollution))
```

```python
# Trying to clean cities names before searching for common cities to extract the most data possible 
# Cities will be foreign key

def normalize_city_name(city):
    if pd.isna(city):
        return None

    city = str(city).lower().strip()
    
    city = unicodedata.normalize('NFKD', city)
    city = ''.join([c for c in city if not unicodedata.combining(c)])
    
    city = city.split(',')[0].strip()
    city = city.split('(')[0].strip()
    
    city = ' '.join(city.split())
    
    return city

df_green['city_clean'] = df_green['City Name'].apply(normalize_city_name)
df_pollution['city_clean'] = df_pollution['city_name'].apply(normalize_city_name)

common_before = len(set(df_green['City Name']) & set(df_pollution['city_name']))
common_after = len(set(df_green['city_clean']) & set(df_pollution['city_clean']))
print(f"Before cleaning: {common_before}")
print(f"After cleaning: {common_after}")
```
